```python
import math
import numpy as np
def relerr(approx,true_val):
    relerr=np.abs(approx-true_val)
    
    if isinstance(approx,(list,np.ndarray)):
        if true_val.any()==0.0:
            nonzero_abs_true_val=[]
            for val in true_val:
                if val==0.0:
                    nonzero_abs_true_val.append(1.0)
                else:
                    nonzero_abs_true_val.append(np.abs(val))
        else:
            nonzero_abs_true_val=np.abs(true_val)
            
        relerr/=nonzero_abs_true_val
    else:
        if true_val !=0.0:
            relerr/=np.abs(true_val)
    return relerr

def _mylog1(x,rtol,atol,max_deg):
    in_x=(x-1.0)/(x+1.0)
    old_ret=in_x
    ret=old_ret
    xn=in_x
    coef=1.0
    for i in range(3,max_deg,2):
        coef=1.0/i
        xn*=in_x**2
        ret=old_ret+coef*xn
        if math.fabs(ret-old_ret)<=rtol*math.fabs(old_ret)+atol:
            return ret*2,i
        old_ret=ret
    return ret*2,i


def _mylog2(x,rtol,atol,max_deg):
    in_x=(x-1.0)/x
    old_ret=in_x
    ret=old_ret
    xn=in_x
    coef=1.0
    for i in range(2,max_deg):
        coef=1.0/i
        xn*=in_x
        ret=old_ret+coef*xn
        if math.fabs(ret-old_ret)<=rtol*math.fabs(old_ret)+atol:
            return ret,i
        old_ret=ret
    return ret,i

def mylog(x,rtol,atol,max_deg,maclaurin_log):
    ret_val,i=np.nan,0
    flag_minus=0
    flag_deg_e=0
    deg_e=0
    in_exp=2.7182818284590452353602874713527
    in_x=x
    if x<0.0:
        return ret_val,i
    if(x>0.0)and(x<1.0):
        in_x=1.0/x
        flag_minus=1
    if in_x>in_exp:
        in_exp_deg=in_exp
        in_exp_degp1=in_exp*in_exp
        for deg in range(1,710):
            if(in_x>in_exp_deg)and(in_x<=in_exp_degp1):
                deg_e=deg
                flag_deg_e=1
                in_x/=in_exp_deg
                break
            in_exp_deg=in_exp_degp1
            in_exp_degp1*=in_exp
    if in_x==1.0:
        ret_val=0.0
    elif (in_x>1.0)and(in_x<in_exp):
        ret_val,i=maclaurin_log(in_x,rtol,atol,max_deg)
    elif in_x==in_exp:
        ret_val=1.0
        
    if flag_deg_e==1:
        ret_val+=deg_e
    if flag_minus==1:
        ret_val=-ret_val
    return ret_val,i

rtol=1.0e-15
atol=0.0
max_deg=1000
x_array=np.linspace(0.1,10,num=10)
maclaurin_val=[0,0]
deg=[0,0]
reldiff=[0,0]
print('    x    ,relerr[0] , relerr[1] , deg[0], deg[1]')
for x in x_array:
    maclaurin_val[0],deg[0]=mylog(x,rtol,atol,max_deg,_mylog1)
    maclaurin_val[1],deg[1]=mylog(x,rtol,atol,max_deg,_mylog2)
    math_val=math.log(x)
    
    reldiff[0]=relerr(maclaurin_val[0],math_val)
    reldiff[1]=relerr(maclaurin_val[1],math_val)
    print(f'{x:10.3e},{reldiff[0]:10.3e},{reldiff[1]:10.3e},{deg[0]:5d},{deg[1]:5d}')
    
# x = 100
x = 100
print('x = %g' % x)
maclaurin_val[0], deg[0] = mylog(x, rtol, atol, max_deg, _mylog1)
maclaurin_val[1], deg[1] = mylog(x, rtol, atol, max_deg, _mylog2)
math_val = math.log(x)  # math.log

reldiff[0] = relerr(maclaurin_val[0], math_val)
reldiff[1] = relerr(maclaurin_val[1], math_val)

print(f'{x:10.3e},{reldiff[0]:10.3e},{reldiff[1]:10.3e},{deg[0]:5d},{deg[1]:5d}')
```
